# **Wind Energy Ireland**



### **Project overview:**

This project analyses the historical weather data from [Met Eireann](https://www.met.ie/climate/available-data/historical-data) for five locations across Ireland representing different regions of the country to get an overview of weather conditions in particular windspeed. 

Locations: 
-   Belmullet, Co. Mayo (1956-2024)
-   Malin Head, Co. Donegal (1955-2024)
-   Dublin Airport, Co. Dublin (1945-2024)
-   Shannon Airport, Co. Clare (1945-2024)
-   Valentia Observatory, Co. Kerry (1945-2024)

### **Background:**

In order to reduce carbon emissions and combat climate change, renewable energy sources such as wind and solar power are becoming an increasingly important part of Irelands energy sources. 
Due to its location on the North Atlantic, Ireland experiences higher than average wind speeds, especially along the west coast which makes it an ideal location for windfarms.  At present, 30% of Irelands electricity generation comes from wind energy and is expected to increase. 
Solar energy contributes a smaller amount of Irelands renewable energy but is expected to grow further as technologies improve. 
The limitations in storage technology means that renewable energy sources such as wind and solar power can be intermittent. However, Ireland aims to meet 70% of its electricity needs from renewable energy sources by 2030. By relying on cleaner energy, Ireland can reduce its carbon emissions and improve air quality.

##### **The Electricity Grid in Ireland:**
The electricity grid in Ireland is a system of power stations, transmission lines and distribution networks that delivers electricity across the country. 
Irelands grid is managed by EirGrid, the national transmission operator and ESB networks. The grid works by balancing the supply and demand of electricity in real-time, which is crucial for maintaining a stable power supply. 
As renewable energy sources like wind and solar power are becoming a larger part of Irelands energy they are increasingly integrated into the grid. However, renewable energy sources can be intermittent and dont always produce energy. To manage this the grid relies on backup power from other sources to ensure a stable supply. Future advancements in storage technology such as batteries, will help to store excess energy when renewable power generation is high and release it when demand increases. However, at present battery storage time is typically 1-2 hours. 

Windfarms generate electricity from wind. They consist of multiple wind turbines that convert the kinetic energy of the wind into electrical energy. These turbines are generally located in areas with high windspeeds.

##### **Wind Energy Belmullet 2019:**

The wind energy data for Belmullet was estimated via the webtool [Renewables Ninja](https://www.renewables.ninja/). This webtool shows the estimated amount of energy that could be generated by wind or solar farms at any location. The data available is for 2019. 

The plot shows a noticeable decrease in energy production during the summer months, particularly from June to August. This is likely due to the lower windspeeds observed in these months, which result in the reduced efficiency of windfarms and therefore lower energy generation at this time. 

There is a strong correlation between windspeed and wind energy. As windspeed increases, wind energy also increases. 

The predicted value of electricity generation was higher in some cases than the actual value. While this data is just an estimate, windspeed alone does not always accurately predict actual electricity generation, as other factors such as turbine efficiency, maintenance downtime etc can play a significant role.

#### **Conclusion:**

Wind energy is the largest renewable energy source in Ireland. Ireland is one of the leading countries in the world for wind energy production due to its high windspeeds and location on the Atlantic Ocean making it an ideal location for onshore and offshore windfarms. The historical data showed a slight decrease but with a slope of 0.000 it is not signficant and the overall trend is that Irelands winds have remained relatively stable. Windspeeds are typically stronger during the winter months. This variability in windspeed means that while wind energy is a valuable source of power, it is not always consistent, and there are periods of low generation that require alternative energy sources. 
Solar energy generation requires sunlight and is most effective during the day and summer months. The variability in both solar and wind energy production could mean that as technologies continue to improve, when one source is less productive the other could generate power giving a more stable energy supply and reduce reliance on fossil fuels.
